```python
teammate_data = (
    merged.groupby(['year','constructorRef','driver_name'])
    .agg(
        Races=('raceId','count'),
        Wins=('positionOrder', lambda x: (x == 1).sum()),
        Podiums=('positionOrder', lambda x: (x <= 3).sum()),
        Points=('points','sum')
    )
    .reset_index()
)

teammate_data.head()
```

```python
teammate_data.shape
```

```python
teammate_data.sort_values(
    ['year','constructorRef']
).head(20)
```

```python
teammate_data = teammate_data[
    teammate_data['Races'] >= 5
]
```

```python
teammate_data.shape
```

```python
team_totals = (
    teammate_data.groupby(['year','constructorRef'])['Points']
    .sum()
    .reset_index(name='Team_Points')
)

teammate_data = teammate_data.merge(
    team_totals,
    on=['year','constructorRef'],
    how='left'
)

teammate_data['Teammate_Dominance'] = (
    teammate_data['Points'] /
    teammate_data['Team_Points']
)

teammate_data.head()
```

```python
teammate_data.sort_values(
    'Teammate_Dominance',
    ascending=False
)[[
    'driver_name',
    'year',
    'constructorRef',
    'Points',
    'Team_Points',
    'Teammate_Dominance'
]].head(20)
```

```python
teammate_filtered = teammate_data[
    (teammate_data['Team_Points'] >= 50) &
    (teammate_data['Races'] >= 5)
].copy()

teammate_filtered.shape
```

```python
teammate_filtered.sort_values(
    'Teammate_Dominance',
    ascending=False
)[[
    'driver_name',
    'year',
    'constructorRef',
    'Points',
    'Team_Points',
    'Teammate_Dominance'
]].head(30)
```

```python
teammate_counts = (
    teammate_filtered.groupby(
        ['year','constructorRef']
    )['driver_name']
    .nunique()
    .reset_index(name='num_drivers')
)

teammate_counts['num_drivers'].describe()
```

```python
teammate_compare = teammate_filtered.copy()

teammate_compare['Best_Teammate_Points'] = (
    teammate_compare.groupby(
        ['year','constructorRef']
    )['Points']
    .transform('max')
)

teammate_compare.head()
```

```python
team_rankings = teammate_filtered.copy()

team_rankings['Point_Rank'] = (
    team_rankings.groupby(
        ['year','constructorRef']
    )['Points']
    .rank(
        method='dense',
        ascending=False
    )
)

team_rankings.sort_values(
    ['year','constructorRef','Points'],
    ascending=[True,True,False]
).head(20)
```

```python
team_rankings[['driver_name','year','constructorRef','Points','Point_Rank']].head(20)
```

```python
teammate_compare = team_rankings.copy()

second_driver = (
    teammate_compare[
        teammate_compare['Point_Rank'] == 2
    ][[
        'year',
        'constructorRef',
        'Points'
    ]]
    .rename(
        columns={
            'Points':'Second_Driver_Points'
        }
    )
)

teammate_compare = teammate_compare.merge(
    second_driver,
    on=['year','constructorRef'],
    how='left'
)

teammate_compare[
    [
        'driver_name',
        'year',
        'constructorRef',
        'Points',
        'Point_Rank',
        'Second_Driver_Points'
    ]
].head(20)
```

```python
top_teammates = teammate_compare[
    teammate_compare['Point_Rank'] <= 2
].copy()

top_teammates.shape
```

```python
head_to_head = top_teammates.copy()

leader_points = (
    head_to_head[
        head_to_head['Point_Rank'] == 1
    ][[
        'year',
        'constructorRef',
        'Points'
    ]]
    .rename(columns={'Points':'Leader_Points'})
)

head_to_head = head_to_head.merge(
    leader_points,
    on=['year','constructorRef'],
    how='left'
)

head_to_head['Head_To_Head_Score'] = (
    head_to_head['Points'] /
    head_to_head['Leader_Points']
)

head_to_head[
    [
        'driver_name',
        'year',
        'constructorRef',
        'Points',
        'Point_Rank',
        'Leader_Points',
        'Head_To_Head_Score'
    ]
].head(20)
```

```python
driver_alpha_teammate = (
    head_to_head.groupby('driver_name')
    .agg(
        Seasons=('year','count'),
        Avg_Teammate_Score=('Head_To_Head_Score','mean')
    )
)

driver_alpha_teammate[
    driver_alpha_teammate['Seasons'] >= 5
].sort_values(
    'Avg_Teammate_Score',
    ascending=False
).head(30)
```
